```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from pprint import pprint

# Import API key
import api_keys
```

## Pull Rankings by Week from API

```python
# Get API Key from local file
your_api_key = api_keys.api_key

ncaamb_week = ['W2', 'W3', 'W4', 'W5', 'W6', 'W7', 'W8', 'W9', 'W10', 'W11', 
               'W12', 'W13', 'W14', 'W15', 'W16', 'W17', 'W18', 'W19', 'W20']

rankings_by_week = []

for week in ncaamb_week:
    
    # Construct url for API call
    access_level = 'trial'
    version = 'v4'
    language_code = 'en'
    poll_name = 'AP'
    season_year = '2018'
    ncaamb_season = 'REG'
    format = 'json'
    aprankingsUrl = f"https://api.sportradar.us/ncaamb/{access_level}/{version}/" + \
        f"{language_code}/polls/{poll_name}/{season_year}/{week}/rankings.{format}?api_key={your_api_key}"


    # Request data for constructing team list from API endpoint
    time.sleep(1)
    response = requests.get(aprankingsUrl).json()
    rankings_by_week.append(response)
    print(f"Pulled JSON from \n{aprankingsUrl}")
```

```python
pprint(rankings_by_week)
```

```python
# Build dataframe from json data pulled
rankings_data = []
for week in rankings_by_week:
    for team in week.get('rankings', ''):
#         print(week['week'])
#         print(week['year'])
#         print(team['wins'])
#         print(team['losses'])
#         print(team['market'])
#         print(team['name'])
#         print(team['rank'])
#         print(' ')    
    
        rankings_dict = {}
        rankings_dict['week'] =  week.get('week', '')
        rankings_dict['id'] =  team['id']
        rankings_dict['mascot'] = team['name']
        rankings_dict['market'] = team['market']
        rankings_dict['rank'] =   team['rank']
        rankings_dict['wins'] =   team['wins']
        rankings_dict['losses'] = team['losses']
        #rankings_dict['prev_rank'] = team['prev_rank'],
        rankings_dict['prev_rank'] = team.get('prev_rank', '')
        rankings_dict['points'] =  team['points']
        
       
        rankings_data.append(rankings_dict)        
```

```python
rankings_data_df = pd.DataFrame(rankings_data)
rankings_data_df = rankings_data_df.set_index('id')
rankings_data_df.head()
```

```python
rankings_data_df = rankings_data_df[['week', 'market', 'mascot', 'rank', 'prev_rank', 'wins', 'losses', 'points']]
rankings_data_df.head()
```

## Save dataframe to csv file

```python
rankings_data_df.to_csv("top_25_rankings_by_week.csv", index=False, header=True)
print(f"Saved data to csv file")
```
