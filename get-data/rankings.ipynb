```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from pprint import pprint

# Import API key
import api_keys
```

## Build Team List

```python
# Get API Key from local file
api_key = 'dzsdr8tp6f3sxcmwuzacchwn'

rankings = [ ]

for ncaamb_week in ['W2','W3','W4', 'W5', 'W6', 'W7', 'W8', 'W9', 'W10', 'W11', 'W12', 'W13', 'W14', 'W15', 'W16', 'W17', 'W18', 'W19']:

    # Construct url for API call
    access_level = 'trial'
    version = 'v4'
    language_code = 'en'
    poll_name = 'AP'
    season_year = '2018'

    format = 'json'
    rankingsbyweekUrl = f"https://api.sportradar.us/ncaamb/{access_level}/{version}/" + \
        f"{language_code}/polls/{poll_name}/{season_year}/{ncaamb_week}/rankings.{format}?api_key={api_key}"
    
    time.sleep(1)

    # Request data for constructing team list from API endpoint
    raw_data = requests.get(rankingsbyweekUrl).json()
    print(f"Pulled JSON from \n{rankingsbyweekUrl}")

    # Build list of team ids to use when calling API

    for ranking in raw_data['rankings']:
        rankings.append({
            "id":          ranking['id'],
            "week": ncaamb_week,
            "school_name": ranking['market'],
            "rank":   ranking['rank'],
            "wins":  ranking['wins'],
            "losses": ranking['losses'],
            "points": ranking['points']
        })

   
pprint(rankings)

```

## Set output file name and save Weeks 2 through 19 in file

```python
# Set name of output data file
output_data_file = "raw-rankings-regseason.txt"
```

```python
with open(output_data_file, 'w') as outfile:  
    json.dump(rankings, outfile)
    print('************************************************')
    print(f"Saved returned data to {output_data_file}")
```
